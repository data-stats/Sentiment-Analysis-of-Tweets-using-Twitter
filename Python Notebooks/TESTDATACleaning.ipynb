**Remove all tagged usernames from tweets as it does not contribute in Sentiment **

```
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt    

df['text'] = np.vectorize(remove_pattern)(df['Tweets'], "@[\w]*") 
```

##Read the csv file 's_last.csv' to pandas dataframe df so as to predict sentiment of these cleaned tweets using TextBlob library

```
df=pd.read_csv('s_last.csv')
```

```
df.head(2)
```

```
type(df['data'])
```

```
del df['Unnamed: 0']
```

```
df.head(1)
```

**Polarity and subjectivity of tweets are predicted and returned in a tuple format to formed a new dataframe column**

```
def sentiment_func(lyrics):
    try:
        return TextBlob(lyrics).sentiment
    except:
        return None
```

```
df['pol_sub'] = df['data'].apply(sentiment_func)
```

```
pd.options.display.max_colwidth = 220
```

```
df
```

**Separate out the Polarity and Subjectivity columns**

```
df[['polarity', 'subjectivity']] = pd.DataFrame(df['pol_sub'].tolist(), index=df.index) 
```
